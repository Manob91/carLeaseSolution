# 3. Business Queries

```python
# Join the Car and Lease Datasets
# Fetch the columns required for the business queries
dfMerge = pd.merge(
  dfLease[["AGREEMENT_ID", "CAR_ID", "RENTAL_MONTH", "DURATION_MONTHS", "RENTAL_START_DATE", "RENTAL_END_DATE", "CAR_RESALE_FLAG", "CAR_RESALE_PRICE"]], 
    dfCars[["CAR_ID", "BRAND", "MODEL_YEAR", "SHOWROOM_PRICE"]],
    how="left",
    left_on=["CAR_ID"], right_on=["CAR_ID"]
)

dfMerge.head()
```

## 3a. Which is the most popular car brand leased in a particular year

```python
# Most popular car brand leased in every year
dfPopularPerYear01 = dfMerge \
.groupby([dfMerge.RENTAL_START_DATE.dt.year, "BRAND"]) \
.AGREEMENT_ID \
.agg(["count"]) \
.reset_index()

# Create RANK column to rank the brands with their popularity each year
dfPopularPerYear01["RANK"] = dfPopularPerYear01 \
.groupby("RENTAL_START_DATE") \
.rank(method="min", ascending=False)

# Filter the most popular brands each year (RANK = 1)
dfPopularPerYear01 \
.loc[dfPopularPerYear01.RANK == 1] \
.sort_values("RENTAL_START_DATE")
```

## 3b. Average lease duration for each car brand

```python
# Average lease duration per brand
dfMerge \
.groupby(["BRAND"]) \
.DURATION_MONTHS \
.agg(["mean"]) \
.reset_index() \
.sort_values(["mean"], ascending=[False])
```

## 3c. Which brand car lease agreements are generally terminated early (hint – Lease end date < (Lease start date + Lease duration in months)

```python
# Rental duration for each agreement
dfMerge01Diff = dfMerge[["AGREEMENT_ID", "BRAND", "DURATION_MONTHS", "RENTAL_START_DATE", "RENTAL_END_DATE"]].assign(
    RENTAL_DURATION = (
        (dfMerge.RENTAL_END_DATE.dt.year - dfMerge.RENTAL_START_DATE.dt.year) * 12
        + dfMerge.RENTAL_END_DATE.dt.month - dfMerge.RENTAL_START_DATE.dt.month
    )
)

# If the rental duration is less than the contract duration, populate the EARLY_TERMINATION_FLAG = "Y"
dfMerge02EarlyTermination = dfMerge01Diff.assign(
    EARY_TERMINATION_FLAG = (
        np.where(dfMerge01Diff.RENTAL_DURATION < dfMerge01Diff.DURATION_MONTHS, "Y", "N")
    )
)

dfMerge02EarlyTermination.head()
```

```python
# Counts of early termination for each brand
dfMerge02EarlyTerminationBrands = dfMerge02EarlyTermination.loc[
    dfMerge02EarlyTermination.EARY_TERMINATION_FLAG == "Y", 
    ["BRAND","EARY_TERMINATION_FLAG"]
] \
.groupby("BRAND") \
.count() \
.reset_index()

dfMerge02EarlyTerminationBrands.sort_values("EARY_TERMINATION_FLAG", ascending=False)
```

## 3d. Keeping aside other factors, how good are the residual value predictions (hint – Compare residual value with the car resale price)

## 3e. Calculate the proft/loss of an agreement

```python
# function to return the difference between two dates in month
def fnDateDiffMonth(startDate, endDate):
    return (endDate.year - startDate.year) * 12 + (endDate.month - startDate.month)

# function to calculate the total revenue for an agreement
# if either of the rental start or end date is equal to 01-Jan-1700 revenue is 0
# if the rental is yet to end, return revenue till date
# if the rental has ended, return the total revenue
# if the car was sold at the end of the rental, add the resale prive to the total rental revenue
def fnTotalRevenue(rentalStart, rentalEnd, monthlyRental, resalePrice, resaleFlag):
    totalRevenue = 0
    durationOfRental = 0
    nullDate = datetime.datetime.strptime("01-Jan-1700", "%d-%b-%Y")
    
    if (rentalStart == nullDate or rentalEnd == nullDate):
        return 0
    if (rentalEnd >= datetime.datetime.now()):
        durationOfRental = fnDateDiffMonth(rentalStart, datetime.datetime.now())
        totalRevenue = durationOfRental * monthlyRental
    elif (rentalEnd < datetime.datetime.now()):
        durationOfRental = fnDateDiffMonth(rentalStart, rentalEnd)
        totalRevenue = durationOfRental * monthlyRental
    
    if resaleFlag == "Y":
        totalRevenue = totalRevenue + resalePrice
    
    return totalRevenue
```

```python
# calculate the total revenue for an agreement using the function
dfMerge["TOTAL_REVENUE"] = dfMerge.apply(
    lambda row: fnTotalRevenue(
        row["RENTAL_START_DATE"],
        row["RENTAL_END_DATE"],
        row["RENTAL_MONTH"],
        row["CAR_RESALE_PRICE"],
        row["CAR_RESALE_FLAG"]
    ),
    axis=1
)

dfMerge.head()
```

```python
# function to get the total revenue for a car and if the car was sold eventually
def fnAggCond(x):
    aggCond = {
        "TOTAL_REVENUE": x["TOTAL_REVENUE"].sum(),
        "RESALE_FLAG": x["CAR_RESALE_FLAG"].max(),
        "RESALE_COUNT": x[x["CAR_RESALE_FLAG"]=="Y"]["CAR_ID"].count(),
    }
    return pd.Series(aggCond)

dfCarRevenue = dfMerge \
.groupby("CAR_ID") \
.apply(fnAggCond) \
.reset_index()

dfCarRevenue.head()
```

```python
# Join with the car data to get the showroom price (buying price)
dfCarResale = pd.merge(
    dfCarRevenue, 
    dfCars[["CAR_ID", "SHOWROOM_PRICE"]],
    how="left",
    left_on=["CAR_ID"], right_on=["CAR_ID"]
)

# If the car was sold, calculate the profit/loss
# If the car is still active for rental, the profit/loss will be 0
dfCarProfitLoss = dfCarResale \
.assign(
    PROFIT_LOSS = (
        np.where(dfCarResale.RESALE_FLAG == "Y", 
                 dfCarResale.TOTAL_REVENUE - dfCarResale.SHOWROOM_PRICE,
                 0 # dfCarResale.TOTAL_REVENUE
                )
    )
).loc[:, ["CAR_ID", "TOTAL_REVENUE", "RESALE_FLAG", "SHOWROOM_PRICE", "PROFIT_LOSS"]]

dfCarProfitLoss.head()
```
